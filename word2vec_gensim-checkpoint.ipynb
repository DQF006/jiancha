```python
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec

# 读入训练集文件
data = pd.read_csv('C:/Users/35866/train.csv')
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
# 词向量模型训练（Skip-Gram：sg=1）
model = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
# 模型显示
print('模型参数（Skip-Gram）：', model, '\n')
#%%
```

```python
# 获取词向量
env_vector = model.wv['环境']
# 输出向量和形状
print('"环境"的词向量：', env_vector, '\n')
print('"环境"词向量形状：', env_vector.shape)
```

```python
# 获取最相似的3个词
top3_similar = model.wv.most_similar('好吃', topn=3)
print('与"好吃"语义最接近的3个词：', top3_similar)
```

```python
# 计算相似度
sim_good_tasty = model.wv.similarity('好吃', '美味')
sim_good_cockroach = model.wv.similarity('好吃', '蟑螂')
print('"好吃"与"美味"的相似度：', sim_good_tasty)
print('"好吃"与"蟑螂"的相似度：', sim_good_cockroach)
```

```python
# 向量运算并取top1
result = model.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1)
print('"餐厅+聚会-安静" 最相关的词：', result[0][0])
```
