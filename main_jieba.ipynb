# 结巴分词实践

## 结巴分词的三种模式

### 全模式

```python
# 导入 jieba 库
import jieba

# 全模式分词：cut_all=True，会扫描句子中所有可能的词语
seg_list = jieba.cut("我来到北京清华大学", cut_all=True)

# 用 "/" 连接分词结果并打印
print("Full Mode: " + "/ ".join(seg_list))
```

### 精确模式

```python
# 精确模式分词：cut_all=False（默认值），会找到最合理的分词结果
seg_list = jieba.cut("我来到北京清华大学", cut_all=False)
print("Default Mode: " + "/ ".join(seg_list))
```

### 搜索引擎模式

```python
# 搜索引擎模式：在精确模式基础上，对长词再次切分，提高召回率
seg_list = jieba.cut_for_search("小明硕士毕业于中国科学院计算所，后在日本京都大学深造")
print(", ".join(seg_list))
```

## 载入词典

```python
# 加载自定义词典
jieba.load_userdict("userdict.txt")

# 测试分词：替换为你的姓名，HMM=False 强制使用词典匹配
seg_list = jieba.cut("杜庆峰即将是自然语言处理方面的高手", cut_all=False, HMM=False)
print("/ ".join(seg_list))
```
