```python
import pandas as pd
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

# ---------------------- 任务1: Skip-gram 模型训练 ----------------------
# 加载数据并预处理
df = pd.read_csv("train.csv")
sentences = [simple_preprocess(text) for text in df["comment"]]  # 这里改成 comment！

# 训练 Skip-gram 模型 (sg=1)
model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1,
    epochs=10
)
print("Skip-gram 模型训练完成")
```

```python
env_vector = model.wv["环境"]
print("词向量（环境）：", env_vector)
print("词向量形状：", env_vector.shape)
```

```python
similar_words = model.wv.most_similar("好吃", topn=3)
print("与'好吃'语义最接近的3个词：", similar_words)
```

```python
sim_delicious_tasty = model.wv.similarity("好吃", "美味")
sim_delicious_cockroach = model.wv.similarity("好吃", "蟑螂")
print("'好吃'与'美味'的相似度：", sim_delicious_tasty)
print("'好吃'与'蟑螂'的相似度：", sim_delicious_cockroach)
```

```python
result = model.wv.most_similar(positive=["好吃", "分量"], negative=["不好吃"], topn=1)
print("好吃 + 分量 - 不好吃 =", result[0][0])
```
